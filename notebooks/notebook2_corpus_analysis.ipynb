## Panjabi-gpt

```python
from pathlib import Path
import re
import json
import math
from collections import Counter

import pandas as pd
```

## path

```python
project_root = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
data_dir = project_root / "data"
outputs_dir = project_root / "outputs"

outputs_dir.mkdir(parents=True, exist_ok=True)

print("Project root:", project_root)
print("Data dir:", data_dir)
print("Outputs dir:", outputs_dir)
```

## find parquet file

```python
parquet_files = sorted(data_dir.glob("sangraha_pan_guru_chunk_*.parquet"))[:2]

print(f"Using {len(parquet_files)} parquet files")
for f in parquet_files:
    print(f.name)
```

## load parquet(dataset) files

```python
dfs = []
for f in parquet_files:
    df = pd.read_parquet(f)
    df["source_file"] = f.name
    dfs.append(df)

corpus_df = pd.concat(dfs, ignore_index=True)

print("Total rows loaded:", len(corpus_df))
print("Columns:", list(corpus_df.columns))
corpus_df.head()
```

## inspect sample row

```python
for i in range(3):
    print(f"\n--- Row {i+1} ---")
    print(corpus_df.iloc[i].to_dict())
```

## check null or duplicate value 

```python
print("Null counts:")
print(corpus_df.isnull().sum())

print("\nDuplicate doc_id count:")
if "doc_id" in corpus_df.columns:
    print(corpus_df["doc_id"].duplicated().sum())
else:
    print("No doc_id column found")
```

## basic cleaning helper text

```python
def clean_text(text: str) -> str:
    if pd.isna(text):
        return ""
    
    text = str(text)

    # Normalize line endings
    text = text.replace("\r\n", "\n").replace("\r", "\n")

    # Remove excessive spaces/tabs
    text = re.sub(r"[ \t]+", " ", text)

    # Remove too many blank lines
    text = re.sub(r"\n{3,}", "\n\n", text)

    # Strip outer whitespace
    text = text.strip()

    return text


def is_gurmukhi_char(ch: str) -> bool:
    # Unicode block for Gurmukhi: U+0A00 to U+0A7F
    return "\u0A00" <= ch <= "\u0A7F"


def gurmukhi_ratio(text: str) -> float:
    if not text:
        return 0.0
    
    chars = [ch for ch in text if not ch.isspace()]
    if not chars:
        return 0.0
    
    gurmukhi_count = sum(1 for ch in chars if is_gurmukhi_char(ch))
    return gurmukhi_count / len(chars)
```

## apply cleaning

```python
corpus_df["text"] = corpus_df["text"].apply(clean_text)

print("After cleaning:")
print(corpus_df["text"].head(3).tolist())
```

## compute text stats

```python
corpus_df["char_len"] = corpus_df["text"].apply(len)
corpus_df["word_len"] = corpus_df["text"].apply(lambda x: len(x.split()))
corpus_df["line_count"] = corpus_df["text"].apply(lambda x: x.count("\n") + 1 if x else 0)
corpus_df["gurmukhi_ratio"] = corpus_df["text"].apply(gurmukhi_ratio)

corpus_df[["char_len", "word_len", "line_count", "gurmukhi_ratio"]].describe()
```

## inspect distribution of bad rows

```python
print("Rows with empty text:", (corpus_df["char_len"] == 0).sum())
print("Rows with char_len < 20:", (corpus_df["char_len"] < 20).sum())
print("Rows with word_len < 3:", (corpus_df["word_len"] < 3).sum())
print("Rows with gurmukhi_ratio < 0.5:", (corpus_df["gurmukhi_ratio"] < 0.5).sum())
```

## preview suspicious row

```python
bad_samples = corpus_df[
    (corpus_df["char_len"] < 20) |
    (corpus_df["word_len"] < 3) |
    (corpus_df["gurmukhi_ratio"] < 0.5)
]

print("Suspicious rows:", len(bad_samples))

for i in range(min(10, len(bad_samples))):
    row = bad_samples.iloc[i]
    print(f"\n--- Suspicious Sample {i+1} ---")
    print("doc_id:", row.get("doc_id", "N/A"))
    print("char_len:", row["char_len"])
    print("word_len:", row["word_len"])
    print("gurmukhi_ratio:", row["gurmukhi_ratio"])
    print("text:", row["text"][:500])
```

## filter corpus

```python
filtered_df = corpus_df[
    (corpus_df["char_len"] >= 30) &
    (corpus_df["word_len"] >= 5) &
    (corpus_df["gurmukhi_ratio"] >= 0.5)
].copy()

print("Original rows:", len(corpus_df))
print("Filtered rows:", len(filtered_df))
print("Removed rows:", len(corpus_df) - len(filtered_df))
```

## remove exact duplicate rows

```python
before_dedup = len(filtered_df)
filtered_df = filtered_df.drop_duplicates(subset=["text"]).copy()
after_dedup = len(filtered_df)

print("Before dedup:", before_dedup)
print("After dedup:", after_dedup)
print("Removed duplicates:", before_dedup - after_dedup)
```

## final corpus stats

```python
total_docs = len(filtered_df)
total_chars = filtered_df["char_len"].sum()
total_words = filtered_df["word_len"].sum()
avg_chars = filtered_df["char_len"].mean()
avg_words = filtered_df["word_len"].mean()

print("Final corpus stats")
print("------------------")
print("Documents:", total_docs)
print("Total characters:", f"{total_chars:,}")
print("Total words:", f"{total_words:,}")
print("Average chars/doc:", round(avg_chars, 2))
print("Average words/doc:", round(avg_words, 2))
```

## approximate token estimation

```python
token_est_3 = total_chars / 3
token_est_4 = total_chars / 4
token_est_5 = total_chars / 5

print("Approx token estimates")
print("----------------------")
print("At 3 chars/token :", f"{int(token_est_3):,}")
print("At 4 chars/token :", f"{int(token_est_4):,}")
print("At 5 chars/token :", f"{int(token_est_5):,}")
```

## inspect clean sample

```python
for i in range(5):
    print(f"\n--- Clean Sample {i+1} ---")
    print(filtered_df.iloc[i]["text"][:1000])
```

## save clean corpus as parquet

```python
cleaned_parquet_path = data_dir / "punjabi_corpus_cleaned.parquet"
filtered_df.to_parquet(cleaned_parquet_path, index=False)

print("Saved cleaned parquet to:", cleaned_parquet_path)
```

## save pretraining text JSONL

```python
cleaned_jsonl_path = data_dir / "punjabi_corpus_cleaned.jsonl"

with open(cleaned_jsonl_path, "w", encoding="utf-8") as f:
    for text in filtered_df["text"]:
        f.write(json.dumps({"text": text}, ensure_ascii=False) + "\n")

print("Saved cleaned JSONL to:", cleaned_jsonl_path)
```

## save plain text corpus too

```python
plain_text_path = data_dir / "punjabi_corpus_cleaned.txt"

with open(plain_text_path, "w", encoding="utf-8") as f:
    for text in filtered_df["text"]:
        f.write(text.replace("\n", " ").strip() + "\n")

print("Saved plain text corpus to:", plain_text_path)
```

## save summary report

```python
summary = {
    "original_rows": int(len(corpus_df)),
    "filtered_rows": int(len(filtered_df)),
    "removed_rows": int(len(corpus_df) - len(filtered_df)),
    "total_characters": int(total_chars),
    "total_words": int(total_words),
    "avg_chars_per_doc": float(avg_chars),
    "avg_words_per_doc": float(avg_words),
    "approx_tokens_3_chars_per_token": int(token_est_3),
    "approx_tokens_4_chars_per_token": int(token_est_4),
    "approx_tokens_5_chars_per_token": int(token_est_5),
}

summary_path = outputs_dir / "corpus_summary.json"

with open(summary_path, "w", encoding="utf-8") as f:
    json.dump(summary, f, ensure_ascii=False, indent=2)

print("Saved summary to:", summary_path)
print(summary)
```
