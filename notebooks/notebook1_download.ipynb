```python
from pathlib import Path
from datasets import load_dataset, Dataset
```

```python
project_root = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
data_dir = project_root / "data"
data_dir.mkdir(parents=True, exist_ok=True)

print("Project root:", project_root)
print("Data dir:", data_dir)
```

```python
dataset_stream = load_dataset(
    "ai4bharat/sangraha",
    name="synthetic",
    split="pan_Guru",
    streaming=True
)

for i, example in enumerate(dataset_stream):
    print(f"\n--- Example {i+1} ---")
    print(example)
    if i >= 2:
        break
```

```python
dataset_stream = load_dataset(
    "ai4bharat/sangraha",
    name="synthetic",
    split="pan_Guru",
    streaming=True
)

chunk_size = 100_000
chunk = []
chunk_id = 0

for i, example in enumerate(dataset_stream, start=1):
    chunk.append(example)

    if len(chunk) >= chunk_size:
        chunk_id += 1
        out_path = data_dir / f"sangraha_pan_guru_chunk_{chunk_id:03d}.parquet"
        Dataset.from_list(chunk).to_parquet(str(out_path))
        print(f"Saved {out_path} with {len(chunk)} rows")
        chunk = []

# Save remainder
if chunk:
    chunk_id += 1
    out_path = data_dir / f"sangraha_pan_guru_chunk_{chunk_id:03d}.parquet"
    Dataset.from_list(chunk).to_parquet(str(out_path))
    print(f"Saved {out_path} with {len(chunk)} rows")
```
