```python
from pathlib import Path
import json
from transformers import AutoTokenizer
```

## paths

```python
project_root = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
data_dir = project_root / "data"
outputs_dir = project_root / "outputs"

text_file = data_dir / "punjabi_corpus_cleaned.txt"
tokenizer_dir = outputs_dir / "punjabi_tokenizer_16k"

outputs_dir.mkdir(parents=True, exist_ok=True)
tokenizer_dir.mkdir(parents=True, exist_ok=True)

print("Text file exists:", text_file.exists(), text_file)
print("Tokenizer output dir:", tokenizer_dir)
```

## corpus preview

```python
with open(text_file, "r", encoding="utf-8") as f:
    for i in range(3):
        line = f.readline().strip()
        print(f"\n--- Sample {i+1} ---")
        print(line[:500])
```

## iterator for tokenizer training

```python
def get_training_corpus(file_path, batch_size=1000):
    batch = []
    with open(file_path, "r", encoding="utf-8") as f:
        for line in f:
            text = line.strip()
            if text:
                batch.append(text)
            if len(batch) >= batch_size:
                yield batch
                batch = []
    if batch:
        yield batch
```

## load a base tokenizer and train new one

```python
base_tokenizer = AutoTokenizer.from_pretrained("gpt2")

new_tokenizer = base_tokenizer.train_new_from_iterator(
    get_training_corpus(text_file),
    vocab_size=16000
)

print("New vocab size:", len(new_tokenizer))
```

## save tokenizer

```python
new_tokenizer.save_pretrained(str(tokenizer_dir))
print("Saved tokenizer to:", tokenizer_dir)
```

## test tokenizer on punjabi text

```python
samples = [
    "ਪੰਜਾਬੀ ਭਾਸ਼ਾ ਲਈ ਇੱਕ ਨਵਾਂ ਭਾਸ਼ਾਈ ਮਾਡਲ ਤਿਆਰ ਕੀਤਾ ਜਾ ਰਿਹਾ ਹੈ।",
    "ਇਹ ਮਾਡਲ ਅਗਲਾ ਟੋਕਨ ਅਨੁਮਾਨ ਲਗਾਉਣ ਦੀ ਕੋਸ਼ਿਸ਼ ਕਰੇਗਾ।",
    "ਸਿਖਲਾਈ ਲਈ ਸਾਫ਼ ਅਤੇ ਉੱਚ-ਗੁਣਵੱਤਾ ਵਾਲਾ ਡਾਟਾ ਬਹੁਤ ਮਹੱਤਵਪੂਰਨ ਹੈ।"
]

for s in samples:
    enc = new_tokenizer(s)
    print("\nTEXT:", s)
    print("TOKENS:", new_tokenizer.convert_ids_to_tokens(enc["input_ids"])[:50])
    print("TOKEN COUNT:", len(enc["input_ids"]))
```

## estimate real token count on a sample of the corpus

```python
sample_limit = 10000
sample_texts = []

with open(text_file, "r", encoding="utf-8") as f:
    for i, line in enumerate(f):
        if i >= sample_limit:
            break
        line = line.strip()
        if line:
            sample_texts.append(line)

token_counts = []
char_counts = []

for text in sample_texts:
    ids = new_tokenizer(text)["input_ids"]
    token_counts.append(len(ids))
    char_counts.append(len(text))

avg_tokens = sum(token_counts) / len(token_counts)
avg_chars = sum(char_counts) / len(char_counts)
chars_per_token = avg_chars / avg_tokens

print("Sample texts:", len(sample_texts))
print("Average chars per sample:", round(avg_chars, 2))
print("Average tokens per sample:", round(avg_tokens, 2))
print("Estimated chars per token:", round(chars_per_token, 3))
```

## estimate full corpus tokens using the trained tokenizer ratio

```python
with open(project_root / "outputs" / "corpus_summary.json", "r", encoding="utf-8") as f:
    summary = json.load(f)

total_characters = summary["total_characters"]
estimated_total_tokens = total_characters / chars_per_token

print("Estimated total tokens with trained tokenizer:", f"{int(estimated_total_tokens):,}")
```

## save tokenizer summary

```python
tokenizer_summary = {
    "vocab_size": len(new_tokenizer),
    "sample_texts_used": len(sample_texts),
    "avg_chars_per_sample": avg_chars,
    "avg_tokens_per_sample": avg_tokens,
    "estimated_chars_per_token": chars_per_token,
    "estimated_total_tokens": int(estimated_total_tokens),
}

summary_path = outputs_dir / "tokenizer_summary.json"
with open(summary_path, "w", encoding="utf-8") as f:
    json.dump(tokenizer_summary, f, ensure_ascii=False, indent=2)

print("Saved tokenizer summary to:", summary_path)
print(tokenizer_summary)
```
